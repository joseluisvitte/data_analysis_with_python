# Analyzing world population data

## Importing python libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Analyzing a database

#### https://datos.bancomundial.org/indicator/SP.POP.TOTL 

```python
dfpob=pd.read_csv('poblacion/API_SP.POP.TOTL_DS2_es_csv_v2_4682121.csv',skiprows=3) #importing data from file
display(dfpob.head()) #showing header
```

## Indexing our data

```python
dfpob.set_index('Country Code', inplace=True) #changing the index to country code
```

```python
dfpob.head() #show the changes on the table
```

## Filtering the data you need

```python
dfpob2=dfpob[[str(k) for k in range(1960,2022)]] #creating a loop to loop through the years in string type
display(dfpob2)
```

## Creating the graphs with the data

```python
countries=['MEX', 'PER', 'COL', 'ESP'] #selecting only four countries
dfpob4=dfpob2.loc[countries] #assigning the new data to a new variable
display(dfpob4)
plt.figure(figsize=(12,6)) #size of figure
for country in countries: #creating a cycle to go through the countries contained in the list
    plt.plot(dfpob4.loc[country], label=country)
plt.title('Comparison of population growth')
plt.xlabel("Ages")
plt.ylabel("Millions")
plt.legend()
plt.grid()
xvalues=np.arange(1960,2022,5)
xvalues=[str(k) for k in xvalues]
plt.xticks(xvalues, rotation='vertical')
plt.show()
```

## Creating the bar graphs with the data

```python
plt.figure(figsize=(12,6))
width_bars=0.35
plt.grid()
bars_index=np.arange(4)
plt.bar(bars_index,dfpob4['1960'], width= width_bars, label='1960')
plt.bar(bars_index + width_bars, dfpob4['2021'], width=width_bars, label='2021')
plt.xticks(bars_index + width_bars, dfpob4.index)
plt.legend()
plt.show()
```

## Creating the pie graphs with the data

#### 1960 data

```python
plt.figure(figsize=(12,6))
plt.title('1960')
plt.pie(dfpob4['1960'], labels=dfpob4.index, shadow=True, autopct='%1.1f%%') #only 1960 data with the countries selected above
plt.show()
```

#### 2021 data

```python
plt.figure(figsize=(12,6))
plt.title('2021')
plt.pie(dfpob4['2021'], labels=dfpob4.index, shadow=True, autopct='%1.1f%%') #only 1960 data with the countries selected above
plt.show()
```
